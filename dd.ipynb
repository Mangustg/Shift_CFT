```python
def random_steps_down(start, target, steps):
    total_drop = start - target  # 62000
    # Генерируем steps−1 случайных уменьшений
    drops = [random.uniform(0, total_drop / steps * 2) for _ in range(steps - 1)]
    # Последнее уменьшение — остаток, чтобы точно дойти до target
    drops.append(total_drop - sum(drops))
    # Строим траекторию
    values = [start]
    for drop in drops:
        values.append(values[-1] - drop)
    return values, drops

# Пример: 100 шагов от 100000 до 38000
path, step_drops = random_steps_down(100000, 38000, 100)
```
